Notebook này tương ứng **Bước 7: Truyền đạt kết quả**.

Mục tiêu là gom insight, kết luận và nội dung có thể đưa vào slide/report.

```python
from pathlib import Path
import sys

import pandas as pd

DEFAULT_ROOT = Path(r"C:\vscode\hoctap\Customer-Churn-Analysis")

def find_project_root(start_path: Path) -> Path:
    for candidate in [start_path, *start_path.parents, DEFAULT_ROOT]:
        if (candidate / 'src').exists() and (candidate / 'configs').exists():
            return candidate
    raise FileNotFoundError('Khong tim thay project root.')

PROJECT_ROOT = find_project_root(Path.cwd().resolve())
metrics_dir = PROJECT_ROOT / 'models' / 'metrics'
trained_models_dir = PROJECT_ROOT / 'models' / 'trained_models'

metrics_path = metrics_dir / 'model_results.csv'
best_summary_path = metrics_dir / 'best_model_summary.csv'
best_model_path = trained_models_dir / 'best_model.pkl'

if metrics_path.exists():
    results_df = pd.read_csv(metrics_path)
else:
    results_df = pd.DataFrame()

if best_summary_path.exists():
    best_summary_df = pd.read_csv(best_summary_path)
else:
    best_summary_df = pd.DataFrame()

print('Model results exists:', metrics_path.exists())
print('Best model summary exists:', best_summary_path.exists())
print('Best model file exists:', best_model_path.exists())
results_df.head()
```

## 1. Insight chính để đưa vào slide

- Churn là khách hàng rời bỏ dịch vụ.
- `Contract`, `tenure`, `MonthlyCharges`, `TechSupport`, `OnlineSecurity`, `PaymentMethod` là các yếu tố đáng chú ý.
- Khách hàng hợp đồng ngắn hạn và thời gian sử dụng thấp có rủi ro churn cao hơn.
- Dịch vụ hỗ trợ như `TechSupport` và `OnlineSecurity` có liên quan đến khả năng giữ chân khách hàng.

## 2. Kết luận kỹ thuật

- Project đi đủ pipeline: thu thập dữ liệu, làm sạch, EDA, modeling, feature selection, evaluation, communication.
- Notebook 07 đọc lại kết quả đã được Notebook 06 lưu trong `models/metrics/`.
- Best model được lưu tại `models/trained_models/best_model.pkl` sau khi chạy Notebook 06.
- Bài toán là binary classification, không phải time series.
- Mô hình được đánh giá bằng accuracy, precision, recall, f1-score và roc_auc.
- Feature importance giúp giải thích vì sao model dự đoán churn.

## 3. Ý nghĩa thực tiễn

- Doanh nghiệp có thể nhận diện sớm khách hàng có nguy cơ rời bỏ.
- Có thể ưu tiên chăm sóc nhóm rủi ro cao.
- Có thể cải thiện dịch vụ hỗ trợ và chính sách hợp đồng.
- Mô hình hỗ trợ quyết định giữ chân khách hàng dựa trên dữ liệu.

## 4. Hướng phát triển

- Tuning sâu hơn bằng `GridSearchCV` hoặc `RandomizedSearchCV`.
- Feature selection nâng cao bằng `RFE` hoặc `RFECV`.
- Giải thích mô hình bằng `SHAP` hoặc `LIME`.
- Triển khai dashboard hoặc web app để dự đoán churn cho khách hàng mới.

```python
if not best_summary_df.empty:
    best = best_summary_df.iloc[0]
    print(f"Best model: {best['best_model']}")
    print(f"Feature strategy: {best['feature_strategy']}")
    print(f"Accuracy: {best['accuracy']:.4f}")
    print(f"Precision: {best['precision']:.4f}")
    print(f"Recall: {best['recall']:.4f}")
    print(f"F1-score: {best['f1']:.4f}")
    print(f"ROC-AUC: {best['roc_auc']:.4f}")
    print(f"Model file: {best['model_path']}")
elif not results_df.empty:
    best = results_df.sort_values('f1', ascending=False).iloc[0]
    print(f"Best model by F1-score: {best['model']}")
    print(f"F1-score: {best['f1']:.4f}")
    print(f"ROC-AUC: {best['roc_auc']:.4f}")
else:
    print('Chua co ket qua. Hay chay Notebook 06 truoc de tao model_results.csv va best_model.pkl.')
```
